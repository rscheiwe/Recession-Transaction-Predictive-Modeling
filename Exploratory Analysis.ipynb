```python
df.groupby('property_type').mean()
```

_Single family homes are more expensive, on average._

```python
sns.boxplot(y = 'property_type', x = 'sqft', data = df)
plt.show()
```

_Single family homes are larger, on average, with a greater variance in distribution._

```python
df.groupby('property_type').agg(['mean', 'std'])
```

```python
correlations = df.corr()

plt.figure(figsize = (7,6))
sns.heatmap(correlations, cmap = 'YlGnBu')
plt.show()
```

```python
mask = np.zeros_like(correlations, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
sns.set_style('white')
plt.figure(figsize = (10,8))

correlate = correlations * 100

sns.heatmap(correlate, annot = True, fmt = '.0f', cmap = 'YlGnBu', mask = mask, cbar = False)
plt.show()
```
